```
#cleaning data
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from wordcloud import WordCloud, STOPWORDS
factory = StopWordRemoverFactory()
hapus = factory.get_stop_words()
sw = list(STOPWORDS)
for i in hapus:
  sw.append(i)

judul = df['Judul']
sentences= []
for s in judul :
  sentences.append(str(s).lower())
print(sentences)

#n-gram
def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]

def gantii(kata_awal, kata_ganti, sentences):
  for sent in sentences:
    for j in range(len(kata_ganti)):
      sent = sent.replace(kata_awal[j], kata_ganti[j])
    sents.append(sent)
    
kata_awal = ['big data', 'machine learning', 'covid 19', 'covid19',
             'small area estimation', 'atima', 'data mining', 'time series',
             'analisis deret waktu', 'data science', 'spasial', 'sains data',
             'analisis data', 'non parametrik', 'kategorik', 'nonparametrik',
             'modelling','pandemi','covid', 'analisa', 'analysis','pendugaan area kecil',
             'regresi',',',':',';','(',')','analisis ', 'model ','data ', 'indonesia',
             'institut','pertanian','bogor', 'method', 'metode '
             ]
kata_ganti = ['bigdata', 'machinelearning', 'covid', 'covid', 
              'sae', 'arima', 'datamining', 'timeseries',
              'timeseries', 'datascience', 'spatial', 'datascience',
              'analisisdata', 'nonparametric-categoic', 'nonparametric-categoric', 'nonparametric-categoric',
              'pemodelan','covid','pandemicovid', 'analisis', 'analisis', 'sae',
              'regression','','','','','', 'analisis','model','data','',
              '','','', 'metode', 'metode'
              ]
sents = []


gantii(kata_awal, kata_ganti, sentences)
print(sents)
vokal=['a','i','u','e','o']
unigram = []
for s in sents:
  unigram.append(get_ngrams(s,1))
print(unigram)
for i in unigram:
  for j in i:
    if j in sw:
      i.remove(j)
#tokenization
grams = unigram

tokens = []
for i in grams:
  for j in i:
    tokens.append(j)
print(tokens)


for i in tokens:
  if i in sw :
   tokens.remove(i)

vowel = ['a','e','i','o','u']

tokens = [word for word in tokens if any(v in word for v in vowel)]
      
print(set(tokens))
len(set(tokens))
tokens=list(set(tokens))
print(tokens)
print(len(tokens))
dftoken=pd.DataFrame(tokens)
dftoken.to_excel('dftoken.xlsx')
```

```
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
xdata = dffix
x_array=np.array(xdata)
scaler = MinMaxScaler()
x_scaled=scaler.fit_transform(x_array)
x_scaled

kmeans=KMeans(n_clusters=5,random_state=123)
kmeans.fit(x_scaled)
print(kmeans.cluster_centers_)
hasil=xdata
hasil["cluster"] = kmeans.labels_+1
hasil["cluster"].to_excel('5clusterall.xlsx')
```
